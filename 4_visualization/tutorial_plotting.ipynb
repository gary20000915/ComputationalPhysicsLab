```python
import numpy as np
import matplotlib.pyplot as plt
```

### prepare some data for visualization

```python
xx = np.linspace(0,5,100) # 0~5 divided 100
y1 = np.sin(xx)
y2 = np.cos(xx)
```

## Line plot

a simplest plot

```python
plt.figure()

plt.subplot(121) # row 1; column 1 and plot figure 1
plt.plot(xx, y1, "--", color = "r", label = "y1")
plt.plot(xx, y2, ":", color = "0.3", label = "y2") # 0.3 (0~1) gray level

plt.subplot(122) # row 1; column 1 and plot figure 2
plt.plot(xx, y1, "--", color = "r", label = "y1")
plt.plot(xx, y2, ":", color = "0.3", label = "y2") # 0.3 (0~1) gray level
plt.legend(loc = "best", frameon = False)
plt.xlabel("x [unit]")
plt.ylabel("y [unit]")
plt.tight_layout()
# plt.title("title_name")
plt.show()
```

## Scatter Plot

```python
plt.figure(1)
plt.scatter(xx, y1, alpha = 1) # alpha: transparency
plt.scatter(xx, y2, alpha = 0.3)
```

# 2D image plots

```python
X, Y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
Z  = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
G  = 1./np.sqrt((X-1)**2+(Y)**2) +1./np.sqrt((X+1)**2+(Y)**2) 
```

```python
plt.figure()
plt.imshow(np.log10(G), extent=[-5,5,-5,5], origin = "lower") # .T is doing a transpose
# origin is in the left upper in computer science; in physics we need to set origin in the left lower.
plt.contour(X, Y, np.log10(G), extent=[-5,5,-5,5], colors = "w",)
plt.xlabel("X")
plt.ylabel("Y")
```
