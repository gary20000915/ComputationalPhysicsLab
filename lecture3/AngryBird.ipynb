```python
import numpy as np
import scipy as sp
import scipy.constants as const
import matplotlib.pyplot as plt
```

```python
dt = 0.001
N = 1000
g = const.g

def AB(PosX, PosY, VelX, VelY):
        X = np.array([PosX])
        Y = np.array([PosY])
        VX = np.array([VelX])
        VY = np.array([VelY])
        for i in range(0, N):
                VY = np.append(VY, VY[i] + -g * dt)
                X = np.append(X, X[i] + VX[0] * dt)
                Y = np.append(Y, Y[i] + VY[i] * dt)
                # print(np.size(X), np.size(Y))
        return [X, Y]

%timeit AB(PosX = 0, PosY = 0, VelX = 10, VelY = 10)

plt.plot(AB(PosX = 0, PosY = 0, VelX = 10, VelY = 10)[0], 
         AB(PosX = 0, PosY = 0, VelX = 10, VelY = 10)[1])
```
